### Q7 — Quartiles & simple outlier rule
For `x = [10,11,12,13,14,50]`, find Q1, Q3, and IQR. Mark values greater than `Q3 + 1.5×IQR` as outliers.

```python
# Your code here
x = np.array([10,11,12,13,14,15,50])

Q1=np.percentile(x,25)
Q3=np.percentile(x,75)

IQR=Q3-Q1

upper_limit = Q3 + 1.5 *IQR

outliers = x[x > upper_limit]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Üst sınır (Q3 + 1.5×IQR):", upper_limit)
print("Aykırı değerler:", outliers)
```

### Q8 — Simple probability
A fair six-sided die is rolled once. Compute the probability of getting an even number.

```python
# Olasılık hesabı: zar atıldığında çift gelme olasılığı

# Tüm olası sonuçlar
outcomes = [1, 2, 3, 4, 5, 6]

# Çift sayılar (istenen sonuçlar)
even_numbers = [x for x in outcomes if x % 2 == 0]

# Olasılık hesabı
probability = len(even_numbers) / len(outcomes)

print("Probability of getting an even number:", probability)
```

### Q10 — Dot product & norm
Let `u = [3,4]`, `v = [1,2]`. Compute the dot product `u·v` and the Euclidean norm of `u`.

```python
import math

u = [3, 4]
v = [1, 2]

# Dot product
dot_product = u[0]*v[0] + u[1]*v[1]

# Euclidean norm of u
norm_u = math.sqrt(u[0]**2 + u[1]**2)

print("Dot product (u·v):", dot_product)
print("Norm of u:", norm_u)
```

### Q12 — Linear regression (tiny)
Generate 50 points: `y = 2x + 1 + noise` with `x ~ U[0,5]`, noise ~ N(0,0.5). Fit `LinearRegression` and print slope, intercept.

```python
from sklearn.linear_model import LinearRegression

# 1. Random seed for reproducibility
np.random.seed(0)

# 2. Generate data
x = np.random.uniform(0, 5, 50)          # 50 points uniformly between 0 and 5
noise = np.random.normal(0, 0.5, 50)     # Gaussian noise with mean 0, std 0.5
y = 2 * x + 1 + noise                    # y = 2x + 1 + noise

# 3. Reshape x for sklearn
x = x.reshape(-1, 1)

# 4. Fit Linear Regression
model = LinearRegression()
model.fit(x, y)

# 5. Print results
print("Slope (coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

### Q13 — Linear regression R²
Using the model from Q12, compute and print the R² score.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 1. Random seed for reproducibility
np.random.seed(0)

# 2. Generate data
x = np.random.uniform(0, 5, 50)
noise = np.random.normal(0, 0.5, 50)
y = 2 * x + 1 + noise
x = x.reshape(-1, 1)

# 3. Fit Linear Regression
model = LinearRegression()
model.fit(x, y)

# 4. Predict y values
y_pred = model.predict(x)

# 5. Compute R² score
r2 = r2_score(y, y_pred)
print("R² score:", r2)
```

### Q14 — Logistic regression (easy)

### Q20 — Simple Neural Network (MLP)
Use `make_moons(n_samples=400, noise=0.25)`. Split 75/25, fit `MLPClassifier(hidden_layer_sizes=(8,))` and print train/test accuracies.

```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# 1. Veri setini oluştur
X, y = make_moons(n_samples=400, noise=0.25, random_state=0)

# 2. Train/test split (%75 - %25)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# 3. Basit MLP modeli oluştur (1 gizli katman, 8 nöron)
mlp = MLPClassifier(hidden_layer_sizes=(8,), max_iter=1000, random_state=0)
mlp.fit(X_train, y_train)

# 4. Train/test tahminleri
y_train_pred = mlp.predict(X_train)
y_test_pred = mlp.predict(X_test)

# 5. Doğruluk hesapla
train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print("Train Accuracy:", train_acc)
print("Test Accuracy:", test_acc)
```
